```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
athlete=pd.read_csv('athlete_events.csv')
region=pd.read_csv('noc_regions.csv')
athlete.head()
```

```python
region.head()
```

```python
# join datasets
athlete_df=athlete.merge(region,how='left',on='NOC')
athlete_df.head(2)
```

```python
athlete_df.shape
```

```python
# column names consistent
athlete_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
athlete_df.head(2)
```

```python
athlete_df.info()
```

```python
athlete_df.isnull().sum()
```

```python
athlete_df.describe()
```

```python
# india details
athlete_df.query('Team=="India"').head()
```

```python
# japan details
athlete_df.query('Team=="Japan"').head()
```

```python
# top countries participating
top10_countries=athlete_df.Team.value_counts().sort_values(ascending=False).head(10)
top10_countries
```

```python
# plot for top 10 countries
plt.figure(figsize=(12,5))
sns.barplot(x=top10_countries.index,y=top10_countries,palette='Set2')
plt.title('Overall participation by countries')
```

```python
# Age Distribution of contries
plt.figure(figsize=(9,5))
plt.title("Age Distribution of Athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athlete_df.Age,bins=np.arange(10,80,2),color='red',edgecolor='white')
plt.show()
```

```python
# winter olympic sports
winter_sports=athlete_df[athlete_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
# summer olympic sports
summer_olympic=athlete_df[athlete_df.Season=='Summer'].Sport.unique()
summer_olympic
```

```python
# male and female participants
gender_counts=athlete_df.Sex.value_counts()
gender_counts
```

```python
# pie chart for male and female athlete
plt.figure(figsize=(9,5))
plt.title('Gender Distribution')
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True)
plt.show()
```
