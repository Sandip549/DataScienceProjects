```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
athlete=pd.read_csv('athlete_events.csv')
region=pd.read_csv('noc_regions.csv')
athlete.head()
```

```python
region.head()
```

```python
# join datasets
athlete_df=athlete.merge(region,how='left',on='NOC')
athlete_df.head(2)
```

```python
athlete_df.shape
```

```python
# column names consistent
athlete_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
athlete_df.head(2)
```

```python
athlete_df.info()
```

```python
athlete_df.isnull().sum()
```

```python
athlete_df.describe()
```

```python
# india details
athlete_df.query('Team=="India"').head()
```

```python
# japan details
athlete_df.query('Team=="Japan"').head()
```

```python
# top countries participating
top10_countries=athlete_df.Team.value_counts().sort_values(ascending=False).head(10)
top10_countries
```

```python
# plot for top 10 countries
plt.figure(figsize=(12,5))
sns.barplot(x=top10_countries.index,y=top10_countries,palette='Set2')
plt.title('Overall participation by countries')
```

```python
# Age Distribution of contries
plt.figure(figsize=(9,5))
plt.title("Age Distribution of Athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athlete_df.Age,bins=np.arange(10,80,2),color='red',edgecolor='white')
plt.show()
```

```python
# winter olympic sports
winter_sports=athlete_df[athlete_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
# summer olympic sports
summer_olympic=athlete_df[athlete_df.Season=='Summer'].Sport.unique()
summer_olympic
```

```python
# male and female participants
gender_counts=athlete_df.Sex.value_counts()
gender_counts
```

```python
# pie chart for male and female athlete
plt.figure(figsize=(9,5))
plt.title('Gender Distribution')
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True)
plt.show()
```

```python
# total medals
athlete_df.Medal.value_counts()
```

```python
# total no. of female athlete in each olympic
female_participant=athlete_df[(athlete_df.Sex=='F') & (athlete_df.Season=='Summer')][['Sex','Year']]
female_participant=female_participant.groupby('Year').count().reset_index()
female_participant.tail()
```

```python
womenOlympics= athlete_df[(athlete_df.Sex=='F') & (athlete_df.Season=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(15,10))
sns.countplot(x='Year',data=womenOlympics,palette='Spectral')
plt.title('Women participation')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(12,8))
part.loc[:,'F'].plot()
plt.title('Plot of Female athletes over time')
```

```python
# Gold medal Athletes
```

```python
GoldMedals = athlete_df[(athlete_df.Medal=='Gold')]
GoldMedals.head()
```

```python
# Take only values that are different from NaN
GoldMedals = GoldMedals[np.isfinite(GoldMedals['Age'])]
```

```python
# Gold beyond 60 years of Age
GoldMedals['ID'][GoldMedals['Age']>60].count()
```

```python
sporting_event = GoldMedals['Sport'][GoldMedals['Age']>60]
sporting_event
```

```python
# plot for sporting event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("GoldMedals of athlete over 60 years age")
```

```python
# Gold medals from each country
GoldMedals = GoldMedals.Region.value_counts().reset_index(name='Medal').head(10)
GoldMedals
```

```python
#GoldMedals.Region.value_counts().reset_index(name='Medal').head(10)
plt.figure(figsize=(15,12))
g=sns.catplot(x='index',y='Medal',data=GoldMedals,height=5,kind='bar',palette='rocket')
g.despine(left=True)
g.set_xlabels("Top 5 contries")
g.set_ylabels('No. of Medals')
plt.title('Gold medals per country')
```

```python
# 2016 Rio Summer Olympics
max_year = athlete_df.Year.max()
print(max_year)
```

```python
team_names = athlete_df[(athlete_df.Year == max_year) & (athlete_df.Medal == 'Gold')].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel("Contrywise medals in year 2016")
```
